```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
import sys
import json
```

```python
files_list=os.listdir('./results/predictions_of_the_models_corrected/') # importing the predictions of the models file names
```

```python
def get_number(file_path): # function to get the number of the pandemic
    return file_path.split('_')[-2] + '_' + file_path.split('_')[-1].split('.')[0]
get_number(files_list[0])
```

```python
def get_code(file_path): # function that returns 'number of day of prediction' + '_' + 'number of pandemic'
    return get_number(file_path) + '_' + file_path.split('_')[1]
get_code(files_list[0])
```

```python
def get_pandemic(file_path ): # get the np.array(n_hospitalized) of the pandemic
    pandemic = pd.read_csv('./all_pandemics/pandemic_'+get_number(file_path)+'.csv')
    pandemic.drop(columns=['Unnamed: 0'],inplace=True)
    return np.array(pandemic.iloc[0])
```

```python
all_stds=dict() 
for i in range(len(files_list)): # building the list of all stds
    df=pd.read_csv('./results/predictions_of_the_models_corrected/'+files_list[i]) 
    df.index=[20*i for i in range(1,df.shape[0]+1)]
    pandemic=get_pandemic(files_list[i])
    df.drop(columns=['Unnamed: 0'],inplace=True)
    df.drop(columns='Exponential', inplace=True)
    df.drop(columns='Exponential Multi', inplace=True) # we dropp the exponential whose prediction explodes
    stds=np.array([np.std(df.iloc[i]) for i in range(df.shape[0])])
    all_stds[get_code(files_list[i])  ] = list(stds)
```

```python
with open('./results/stds_of_the_models_corrected.json', 'w') as f: # writing the results
    json.dump(all_stds, f)
```

```python
with open('./results/stds_of_the_models_corrected.json', 'r') as f: # reading the file containing all stds list
    all_stds = json.load(f)
all_stds
```
